```python
import numpy as np
import pandas as pd
```

```python
prices = pd.read_csv("prices.txt", delim_whitespace=True, header=None)
```

```python
corr_matrix = prices[0:175].corr()
```

```python
# Convert correlation matrix into sorted dictionary

corr_dict = {}
for i in range(corr_matrix.shape[0]):
    single_stock = {}
    for j in range(len(corr_matrix[i])):
        if (i != j):
            single_stock[j] = corr_matrix[i][j]
    single_stock = sorted(single_stock.items(), key=lambda x:abs(x[1]))
    corr_dict["Stock " + str(i)] = single_stock
```
