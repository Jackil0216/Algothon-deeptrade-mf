```python
import json
import pandas as pd
```

```python
with open('../data/stock_std_above_1.json', 'r') as f:
    stocks_std_above_1 = json.load(f)

with open('../data/stock_std_above_1_5.json', 'r') as f:
    stocks_std_above_1_5 = json.load(f)

with open('../data/stock_std_below_1.json', 'r') as f:
    stocks_std_below_1 = json.load(f)

with open('../data/stock_std_below_1_5.json', 'r') as f:
    stocks_std_below_1_5 = json.load(f)
```

```python
curated_data = pd.read_csv('../data/curated_data.csv')
```

```python
data_std_above_1 = curated_data[curated_data['stock'].isin(stocks_std_above_1)]
data_std_above_1_5 = curated_data[curated_data['stock'].isin(stocks_std_above_1_5)]
data_std_below_1 = curated_data[curated_data['stock'].isin(stocks_std_below_1)]
data_std_below_1_5 = curated_data[curated_data['stock'].isin(stocks_std_below_1_5)]
```
