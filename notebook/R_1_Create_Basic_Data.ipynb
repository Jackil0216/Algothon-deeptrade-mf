```python
import pandas as pd
```

```python
def loadPricesAsDF(fn):
    global nt, nInst
    #df=pd.read_csv(fn, sep='\s+', names=cols, header=None, index_col=0)
    df=pd.read_csv(fn, sep='\s+', header=None, index_col=None)
    nt, nInst = df.values.shape
    return df

pricesFile="../prices.txt"
prcAll = loadPricesAsDF(pricesFile)
```

```python
prcAll
```

```python
# create data

closePrice = []
for stock in range(50):
    closePrice.extend(prcAll[stock])

dates = []
for stock in range(50):
    for day in range(250):
        dates.append(day)

stocks = []

for stock in range(50):
    stocks.extend([stock]*250)


full_data = pd.DataFrame({'date': dates, 'stock': stocks, 'closePrice': closePrice})
```

```python
full_data.to_csv('../data/basic_data.csv', index=False)
```
